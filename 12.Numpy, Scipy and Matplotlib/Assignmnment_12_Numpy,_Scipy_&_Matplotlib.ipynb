```python
import numpy as np
import matplotlib.pyplot as plt
a= np.linspace(0,5*np.pi, 10)  #This line creates an array n using NumPy's linspace function.
sin = np.sin(a)   #This line calculates the sine of each element in the array a using NumPy's sin function.
plt.plot(a,sin)    #This line plots the sine wave.
plt.xlabel("time")  #This line adds a label to the x-axis.
plt.ylabel("amplitude")  #This line adds a label to the y-axis.
plt.title("Sin wave")  #This line adds a title to the plot.
plt.xlim(0,5*np.pi)  #This line sets the limits of the x-axis to show values between 0 and 5 * Ï€.
plt.ylim(-1.5,1.5)  #This line sets the limits of the y-axis to show values between -1.5 and 1.5.
plt.show() #This line displays the plot.
```

```python
import numpy as np
from scipy.linalg import lu
a = np.mat("1,2;3,4;5,6") #Creates a 3x2 matrix 'a' using NumPy.
print(a)

a_transpose = np.transpose(a) #Transposes matrix 'a' to create a new matrix 'aT'.
print(a_transpose)

b = 2*np.eye(3) #Creates a 3x3 identity matrix 'b' with all diagonal elements set to 2.
print(b)

c= b*a #Multiplies matrices 'b' and 'a' to create a new matrix 'c'.
print(c)

lu_decomposition =lu(a) #Performs LU decomposition on matrix 'a' using the lu() function from SciPy
print(lu_decomposition)
```
